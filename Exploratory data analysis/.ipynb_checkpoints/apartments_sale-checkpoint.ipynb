В датасете очень много пропущенных значений
Больше всего пропусков в столбцах с названиями: is_apartment, balcony, parks_nearest, ponds_nearest
То есть отсутствует информация о ближайших парках, водоемах, количестве балконов и о том, является ли квартира апартаментами или нет

Остановимся подробнее на пропущенных значениях

- ***населенный пункт*** - скорее всего просто невнимательность, т.к. в этом столбце меньше всего пропусков. Либо, как вариант, названия этих населенных пунктов(например, глубинки) отсутствуют на сервере данной площадке
- ***общее кол-во этажей*** - при наличии большого кол-ва этажей в здании, человек может не знать эту информацию, либо просто лень посмотреть документацию
- ***площадь кухни, общая жилая площадь*** - в отсутствии тех. документации, человек может не знать эти параметры
- ***срок размещения объявления*** - могут быть проблемы как со стороны площадки, так и со стороны людей, которые вовремя не снимают объявления после успешной сделки
- ***парки, пруды и аэропорты рядом или в диапазоне 3км*** могут просто отсутствовать, поэтому люди не указывают эту информацию
- ***высота потолков*** - параметр, который знают не все, и измерить его без инструмента довольно сложно
- ***наличие балкона*** - балкон так же может отсутствовать в квартире
- ***апартаменты*** - возможно, не все понимают значение этого слова, поэтому не могут определить принадлежность своего жилья к ним

Заменим отсутствующие значения на нули там, где это возможно


```python
# Заменяем значения на нули там, где это позволяет сделать логика
# Проверяем, значения поменялись

data['ceiling_height_m'].fillna(data['ceiling_height_m'].mean(), inplace = True)
data['balcony'] = data['balcony'].fillna(0)
data['parks_around_3km'] = data['parks_around_3km'].fillna(0)
data['ponds_around_3km'] = data['ponds_around_3km'].fillna(0)
data['is_apartment'] = data['is_apartment'].fillna(0)
data.isna().sum().sort_values()
```

```python
# Удалим строки без города, они будут мешать в анализе и даже пользователям такие объявления не нужны,
# т.к. отсутствует базовая информация о локации квартиры

data = data.dropna(subset=['locality_name'])

data.info()
```

# Изменение типов данных

Посмотрим, в каких столбцах требуется изменить тип данных

- first_day_exposition - дата публикации должна иметь формат datetime64
- floors_total - кол-во этажей должно быть целым числом
- balcony - кол-во балконов должно быть целым числом
- days_exposition - кол-во дней тоже должно быть целым числом

Изменим данные и посмотрим что получится

```python
# Заменим пропущенные значения в этих столбцах на "unknown", чтобы изменить тип данных

data['floors_total'] = data['floors_total'].fillna(0)
data['balcony'] = np.floor(pd.to_numeric(data['balcony'], errors='coerce')).astype('Int64')   
```

```python
# Меняем типы нужных данных и смотрим что получилось

data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format="%Y-%m-%dT%H:%M:%S")
data['floors_total'] = data['floors_total'].astype('Float64').astype('Int64')
data['balcony'] = data['balcony'].astype('Float64').astype('Int64')
data['days_exposition'] = data['days_exposition'].astype('Float64').astype('Int64')
data.info()
```

**Типы данных успешно изменены**

# Добавляем в таблицу новые столбцы

В наш датасет нужно добавить новые столбцы со следующими параметрами:

- цена одного квадратного метра
- день недели публикации объявления (0 — понедельник, 1 — вторник и так далее)
- месяц публикации объявления
- год публикации объявления

Добавим столбец с ценой за один м2

```python
# Делим столбец с общей ценой на столбец с общей площадью в м2

data['price_per_m2'] = data['last_price'] / data['total_area_m2']
data['price_per_m2'] = data['price_per_m2'].round(decimals=2)
```

```python
data.head()
```

Теперь добавим столбцы с днем недели, месяцем и годом публикации объявления

```python
data['day_of_week_exposition'] = data['first_day_exposition'].dt.day_name()
data['month_exposition'] = data['first_day_exposition'].dt.month_name()
data['year_exposition'] = data['first_day_exposition'].dt.year
```

```python
data.head()
```

Добавим столбец с этажами

```python
# Создадим функцию, которая будет классифицировать этажи

def floor_class(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
```

```python
data['floor_class'] = data.apply(floor_class, axis=1)
```

```python
# Найдем процентное соотношение объявлений в Петербурге и за его пределами
# Отношение получается: 66% Петербург, остальное за ним

(data['locality_name'].value_counts() / data.shape[0] * 100).head(11)
```

Будем считать, что все, находящееся за Санкт-Петербургом - пригород

Посмотрим на топ-20 по стоимости квадратного метра.

```python
data.groupby('locality_name').agg({'price_per_m2':'mean'}).sort_values(
    by='price_per_m2', ascending=False).head(20).round(2)
```

```python
# Создадим фунцкию, которая будет фильтровать объявления 
# Так же создадим отдельный столбец data['town_class'], в котором будет информация где находится квартира:
# в СПб или за его пределами

def locality_class(row):
    if row['locality_name'] == 'Санкт-Петербург':
        return 'СПб'
    else:
        return 'За СПб'
```

```python
# Добавляем все, что сделала функция в новый столбец

data['locality_class'] = data.apply(locality_class, axis=1)
```

```python
# Создаем топ-10 населенных пунктов согласно нашему заданию

top_10_data = data.groupby('locality_name')\
    .agg({'locality_class':'count','price_per_m2':'mean'})
top_10_data = top_10_data.sort_values('locality_class', ascending=False).head(10)
```

```python
# Смотрим, что получилось

top_10_data.sort_values('locality_class', ascending=False).head(10).round(2)
```

На выходе имеем таблицу с ценами за один м2 в 10 населённых пунктах с наибольшим числом объявлений. Лидирует Санкт-Петербург, закрывает десятку лидеров Выборг.

```python
# Создадим сводную таблицу по нашему топ-10 и посмотрим на средние значения в каждом населенном пункте

top_10_data.pivot_table(index='locality_name', 
                        values=['locality_class','price_per_m2'],
                        aggfunc='mean').sort_values(by='locality_class', ascending=False).round(2)
```

```python
# Далее построим график, чтобы выделить населенные пункты с самой высокой и самой низкой стоимостью м2

top_10_data.sort_values('price_per_m2', ascending=True)\
    .drop('locality_class', axis=1).plot(kind='barh');
```

Дороже всего недвижимость стоит в Санкт-Петербурге (неожиданно), а самые низкие цены по нашей выборке в Выборге.

- в Санкт-Петербурге составляет **114849.01**

Напрашивается очевидный вывод: квартиры в центре стоят дороже, чем квартиры за ним. Идет вполне логичная восходящая тенденция: чем ближе к центру, тем дороже недвижимость. Резкий пик в радиусе 0 - 1 км  от центра города, там цена за один квадратный метр доходит до 250 тысяч рублей. На расстоянии 30 км от центра города, жилье стоит почти в 5 раз дешевле.
