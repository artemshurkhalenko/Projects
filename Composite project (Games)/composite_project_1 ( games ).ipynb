## Проверка гипотез

Требуется проверить гипотезы:
- Средние пользовательские рейтинги платформ **Xbox One** и **PC** одинаковые;
- Средние пользовательские рейтинги жанров **Action** (англ. «действие», экшен-игры) и **Sports** (англ. «спортивные соревнования») разные.

Пороговое значение **alpha** задаем самостоятельно

Пояснить:
- Как мы сформулировали нулевую и альтернативную гипотезы;
- Какой критерий применили для проверки гипотез и почему.

```python
df_actual[df_actual['rating'] == -1]
```

### Гипотеза №1

Будем использовать t-критерий Стьюдента, так как сравниваемые выборки независимы между собой

Пороговое значение **alpha** устанавливаем 0.5

Средние пользовательские рейтинги платформ **Xbox One** и **PC** одинаковые

- **Н0** (нулевая гипотеза) - средние пользовательские рейтинги платформ **Xbox One** и **PC** одинаковые.

- **Н1** (альтернативная гипотеза) - средние пользовательские рейтинги платформ **Xbox One** и **PC** отличаются между собой.

```python
Xbox_one_rating = df_actual.query('platform == "XOne" and user_score > 0')['user_score']

PC_rating = df_actual.query('platform == "PC" and user_score > 0')['user_score']

alpha = .05 

results = st.ttest_ind(
    Xbox_one_rating, 
    PC_rating)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Мы не можеет отвергнуть нулевую гипотезу при пороговом значении **alpha 0.05**. Но и утверждать, что пользовательские рейтинги платформ **Xbox One** и **PC** отличаются между собой мы тоже не можем

### Гипотеза №2

Средние пользовательские рейтинги жанров **Action** и **Sports** разные

- **Н0** (нулевая гипотеза) - средние пользовательские рейтинги жанров **Action** и **Sports** одинаковые.
- **Н1** (альтернативная гипотеза) - средние пользовательские рейтинги жанров **Action** и **Sports** отличаются между собой.

```python
Action_rating = df_actual.query('genre == "Action" and user_score > 0')['user_score']

Sports_rating = df_actual.query('genre == "Sports" and user_score > 0')['user_score']

alpha = .05 

results = st.ttest_ind(
    Action_rating, 
    Sports_rating)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Мы отвергаем нулевую гипотезу при пороговом значении **alpha 0.05** и не можем утверждать, что средние пользовательские рейтинги жанров **Action** и **Sports** одинаковые

**Промежуточный вывод:**

Нулевая гипотеза (**H0**) - принимаемое по умолчанию предположение о том, что значения сравниваемых событий равны

Альтернативная гипотеза (**H1**) - это утверждение, которое принимается верным, если отбрасывается **H0**.Такая альтернативная гипотеза называется двусторонней, потому что есть возможное отклонение и в большую, и в меньшую стороны. Альтернативная гипотеза противоречит нулевой гипотезе, — либо нулевая гипотеза не отвергается, либо отвергается и принимается **H1**

В первом случае гипотезы мы сформулировали так:

- **Н0** (нулевая гипотеза) - средние пользовательские рейтинги платформ **Xbox One** и **PC** одинаковые.
- **Н1** (альтернативная гипотеза) - средние пользовательские рейтинги платформ **Xbox One** и **PC** отличаются между собой.

Во стором случае так:

- **Н0** (нулевая гипотеза) - средние пользовательские рейтинги жанров **Action** и **Sports** одинаковые.

Для проверки гипотез мы использовали t-критерий Стьюдента, так как нам нужно сравнить между собой две независимые подборки
